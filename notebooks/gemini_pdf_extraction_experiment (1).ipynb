```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-pro",
    temperature=0
)
```

```python
from PyPDF2 import PdfReader

pdf_path = "Auto-Loan-Agreement-Template.pdf"
reader = PdfReader(pdf_path)

contract_text = ""
for page in reader.pages:
    contract_text += page.extract_text() or ""

print(contract_text[:1500])
```

```python
prompt = f"""
You are a car finance expert.

From the following auto loan agreement, extract and return ONLY valid JSON
with these fields (use null if missing):

- loan_amount
- monthly_payment
- interest_rate_apr
- late_fee_clause
- prepayment_clause
- penalty_or_remedy_clause

Contract text:
{contract_text}
"""

response = llm.invoke(prompt)
print(response.content)
```

```python
from google import genai
client = genai.Client(api_key="YOUR_API_KEY")
```

```python
from PyPDF2 import PdfReader

reader = PdfReader("Auto-Loan-Agreement-Template.pdf")
contract_text = ""

for page in reader.pages:
    contract_text += page.extract_text() or ""

print(contract_text[:800])
```

```python
prompt = f"""
You are a car finance expert.

From the following auto loan agreement, extract and return ONLY valid JSON
with these fields (use null if missing):

- loan_amount
- monthly_payment
- interest_rate_apr
- late_fee_clause
- prepayment_clause
- penalty_or_remedy_clause

Contract text:
{contract_text}
"""

response = client.models.generate_content(
    model="models/gemini-1.5-flash",
    contents=prompt
)

print(response.text)
```
