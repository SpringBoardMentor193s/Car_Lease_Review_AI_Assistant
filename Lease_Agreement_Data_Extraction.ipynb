```python
import pdfplumber

pdf_path = list(uploaded.keys())[0]
print("Reading file:", pdf_path)

all_text = ""
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            all_text += text + "\n"

print(all_text[:500])
```

```python
import re

def extract_lease_fields(text):
    data = {}

    # VIN (17 characters)
    vin = re.search(r'\b[A-HJ-NPR-Z0-9]{17}\b', text)
    data["VIN"] = vin.group() if vin else "Not found"

    # Lease Term (months)
    lease_term = re.search(r'(\d+)\s+months', text, re.IGNORECASE)
    data["Lease_Term"] = lease_term.group() if lease_term else "Not found"

    # Monthly Payment
    payment = re.search(r'\$\s?\d+', text)
    data["Monthly_Payment"] = payment.group() if payment else "Not found"

    # APR / Interest Rate
    apr = re.search(r'Annual Percentage Rate\s*%?\s*(\d+(\.\d+)?)', text, re.IGNORECASE)
    data["APR"] = apr.group(1) if apr else "Not found"

    # Residual Value / Buyout
    residual = re.search(r'Purchase Option Price.*?\$\s?\d+', text, re.IGNORECASE)
    data["Residual_Value"] = residual.group() if residual else "Not found"

    # Mileage Limit
    mileage = re.search(r'(\d{2,6})\s+kilometers|\d{2,6}\s+miles', text, re.IGNORECASE)
    data["Mileage_Limit"] = mileage.group() if mileage else "Not found"

    # Late Fee
    late_fee = re.search(r'late.*?(\$\s?\d+)', text, re.IGNORECASE)
    data["Late_Fee"] = late_fee.group(1) if late_fee else "Not found"

    return data


result = extract_lease_fields(all_text)
for key, value in result.items():
    print(f"{key}: {value}")

```
