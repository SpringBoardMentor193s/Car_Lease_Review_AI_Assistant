```python
from google.colab import files
uploaded = files.upload()
import os

pdf_path = list(uploaded.keys())[0]
print("Reading file:", pdf_path)
import pdfplumber

all_text = ""

with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        page_text = page.extract_text()
        if page_text:
            all_text += page_text + "\n"

print(all_text[:1000])
import re

def extract_details(text):
    details = {}

    # VIN (17 characters)
    vin = re.search(r'\b[A-HJ-NPR-Z0-9]{17}\b', text)
    details["VIN"] = vin.group() if vin else "Not found"

    # Car Model & Make (simple heuristic)
    model = re.search(r'(Toyota|Honda|Hyundai|Maruti|Tata|Ford|BMW|Audi)\s+\w+', text, re.IGNORECASE)
    details["Car_Model"] = model.group() if model else "Not found"

    # Year
    year = re.search(r'\b(20[0-2][0-9])\b', text)
    details["Year"] = year.group() if year else "Not found"

    # Interest Rate
    interest = re.search(r'(\d+(\.\d+)?\s*%)', text)
    details["Interest_Rate"] = interest.group() if interest else "Not found"

    # Monthly Payment
    payment = re.search(r'₹\s?\d{3,6}', text)
    details["Monthly_Payment"] = payment.group() if payment else "Not found"

    return details


extracted_details = extract_details(all_text)
extracted_details

```

```python
from google.colab import files
uploaded = files.upload()

import cv2
import pytesseract
import re

image_path = list(uploaded.keys())[0]
print("Reading image:", image_path)

img = cv2.imread(image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

extracted_text = pytesseract.image_to_string(gray)

print("---- OCR TEXT ----")
print(extracted_text)

print("\nOCR successful, extracted_text created.")
print("OCR TEXT PREVIEW:")
print(extracted_text)


def extract_vin(text):
    text = text.replace(" ", "").replace("\n", "")
    vin_pattern = r'[A-HJ-NPR-Z0-9]{17}'
    match = re.search(vin_pattern, text)
    return match.group() if match else None
```

```python
vin_number = extract_vin(extracted_text)
print("Extracted VIN:", vin_number)
```

```python

vin_number = extract_vin(extracted_text)
print("OCR-detected VIN:", vin_number)


if not vin_number:
    vin_number = input("Enter VIN manually for lookup: ")

print("Final VIN used:", vin_number)
```

```python
import requests

def get_car_details_from_vin(vin):
    url = f"https://vpic.nhtsa.dot.gov/api/vehicles/DecodeVin/{vin}?format=json"
    response = requests.get(url)
    data = response.json()

    car_details = {}
    for item in data['Results']:
        if item['Value']:
            car_details[item['Variable']] = item['Value']

    return car_details


if vin_number:
    car_details = get_car_details_from_vin(vin_number)

    print("---- VEHICLE DETAILS ----")
    for key, value in car_details.items():
        print(f"{key}: {value}")
else:
    print("VIN not found.")
```
